```python
import os
from collections import defaultdict

file_paths = [
    '/content/drive/MyDrive/Research/Anti-Viral_Peptides/Version 4.0 (Review 2)/Code/test_neg.fasta',
    '/content/drive/MyDrive/Research/Anti-Viral_Peptides/Version 4.0 (Review 2)/Code/test_positive.fasta',
    '/content/drive/MyDrive/Research/Anti-Viral_Peptides/Version 4.0 (Review 2)/Code/train_neg.fasta',
    '/content/drive/MyDrive/Research/Anti-Viral_Peptides/Version 4.0 (Review 2)/Code/train_positive.fasta'
]

file_contents = {}
for file_path in file_paths:
    if os.path.exists(file_path):
        with open(file_path, 'r') as f:
            file_contents[file_path] = f.read()
    else:
        print(f"File not found: {file_path}")

# Check for duplicate files based on content
content_to_files = defaultdict(list)
for file_path, content in file_contents.items():
    content_to_files[content].append(file_path)

duplicate_files = {content: files for content, files in content_to_files.items() if len(files) > 1}

if duplicate_files:
    print("Duplicate files found (based on content):")
    for content, files in duplicate_files.items():
        print(f"  Content shared by: {', '.join(files)}")
else:
    print("No duplicate files found.")
```

```python
import os
import re

directory_path = '/content/drive/MyDrive/Research/Anti-Viral_Peptides/Version 4.0 (Review 2)/Code/'
loaded_files = [os.path.basename(path) for path in file_paths]

all_files_in_directory = os.listdir(directory_path)

# Define a regex pattern to match train and test files (assuming they end with _neg.fasta or _positive.fasta)
# This pattern can be adjusted based on the actual naming convention if it's different
train_test_pattern = re.compile(r'(train|test)_(neg|positive)\.fasta$')

remaining_files = []
for file_name in all_files_in_directory:
    if train_test_pattern.search(file_name) and file_name not in loaded_files:
        remaining_files.append(file_name)

if remaining_files:
    print("The following files matching the train/test pattern were found but not loaded:")
    for file_name in remaining_files:
        print(f"- {file_name}")
else:
    print("No additional files matching the train/test pattern were found in the directory.")
```
